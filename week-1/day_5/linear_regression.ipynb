```python
import pandas as pd
data = pd.read_csv('http://raw.githubusercontent.com/ruslantur77/ml-journey/refs/heads/main/day_5/house_prices.csv')
```

```python
data
```

```python
unscaled_X = data.drop('median_house_value', axis=1)
unscaled_y = data['median_house_value']
```

```python
unscaled_X
```

```python
from sklearn.preprocessing import StandardScaler
X_scale = StandardScaler()
y_scale = StandardScaler()
```

```python
X = X_scale.fit_transform(unscaled_X)
y = y_scale.fit_transform(unscaled_y.values.reshape(-1, 1)).ravel()
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
prediction
```

```python
scaled_prediction = y_scale.inverse_transform(prediction.reshape(-1, 1)).ravel()
```

```python
scaled_prediction
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_scale.inverse_transform(y_test.reshape(-1, 1)).ravel(), scaled_prediction)
```

```python
mean_squared_error(y_scale.inverse_transform(y_test.reshape(-1, 1)).ravel(), scaled_prediction) ** 0.5
```
