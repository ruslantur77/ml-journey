```python
from sklearn.datasets import make_classification #создаем синтетический датасет
X, y = make_classification(
    n_samples=1000,        # сколько объектов (строк) создать
    n_features=20,         # сколько всего признаков (колонок)
    n_informative=15,      # из них «полезных» (действительно влияют на y)
    n_redundant=5,         # остальные «лишние» (линейные комбинации полезных)
    weights=[0.95, 0.05],  # доля классов (95 % «0», 5 % «1»)
    random_state=42        # чтобы результат был воспроизводимый
)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,
                                                    stratify=y)  #stratify гарантирует, что распределение классов (меток y) будет одинаковым в обучающей и тестовой выборке.
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
print("accuracy:", accuracy_score(y_test, y_pred))          # доля правильных ответов
print("precision:", precision_score(y_test, y_pred))        # точность (TP / (TP + FP))
print("recall:", recall_score(y_test, y_pred))              # полнота (TP / (TP + FN))
print("f1:", f1_score(y_test, y_pred))                      # F1 = 2·(precision·recall)/(precision+recall)
print("roc-auc:", roc_auc_score(y_test, clf.predict_proba(X_test)[:,1]))  # площадь под ROC-кривой
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_bal, y_bal = sm.fit_resample(X_train, y_train)
```

SMOTE – Synthetic Minority Over-sampling Technique
создаёт синтетические примеры миноритарного класса, интерполируя между реальными объектами.

Что происходит внутри `SMOTE.fit_resample(X_train, y_train)`:

1. SMOTE находит все объекты миноритарного класса (у нас – «1»).  
2. Для каждого такого объекта выбирает k=5 ближайших «соседей» того же класса.  
3. Случайно выбирает одного из этих соседей и **рисует отрезок** между двумя точками в признаковом пространстве.  
4. На этом отрезке случайно ставит новую точку – **синтетический пример**.  
5. Повторяет процесс, пока количество «1» не станет равным количеству «0».

Итог: `X_bal, y_bal` содержат исходные данные плюс новые «фальшивые» строки, которые очень похожи на реальные «1», но не дублируют их дословно.

• Столбцы не меняются — порядок признаков остаётся тем же, что был во входном X_train.

• Значения X меняются — добавляются новые строки (синтетические объекты), но внутри каждой строки признаки интерполируются между двумя реальными соседями, поэтому каждое отдельное числовое значение в новой строке может слегка отличаться от исходных.

```python
clf_smote = RandomForestClassifier(random_state=42)
clf_smote.fit(X_bal, y_bal)
```

```python
import numpy as np
print('единицы:')
print(np.size(y))
print(np.count_nonzero(y==1))
print(np.size(y_bal))
print(np.count_nonzero(y_bal==1))
print()
print('нули:')
print(np.count_nonzero(y==0))
print(np.count_nonzero(y_bal==0))
```

Было 1000 объектов: 950 «0» и 50 «1».

После SMOTE стало 1326 объектов: 663 «0» и 663 «1» (равные классы).
То есть SMOTE добавил 613 новых «1», не трогая «0».

```python
y_pred_smote = clf_smote.predict(X_test)
```

```python
print("accuracy:", accuracy_score(y_test, y_pred_smote))          # доля правильных ответов
print("precision:", precision_score(y_test, y_pred_smote))        # точность (TP / (TP + FP))
print("recall:", recall_score(y_test, y_pred_smote))              # полнота (TP / (TP + FN))
print("f1:", f1_score(y_test, y_pred_smote))                      # F1 = 2·(precision·recall)/(precision+recall)
print("roc-auc:", roc_auc_score(y_pred_smote, clf.predict_proba(X_test)[:,1]))  # площадь под ROC-кривой
```

Итоги:

• Сгенерировали синтетический датасет (95 % «0», 5 % «1»).

• Обучили RandomForest без балансировки → Recall 6 %.

• Применили SMOTE для синтетического увеличения миноритарного класса.

• Получили рост Recall до 31 % и F1 до 0.37, показав, как бороться с дисбалансом классов.



Метрики (коротко)

accuracy – доля правильных ответов.

precision – сколько из предсказанных «1» действительно «1».

recall – сколько реальных «1» мы нашли.

f1 – гармоническое среднее precision и recall.

roc-auc – площадь под ROC-кривой (способность различать классы).
